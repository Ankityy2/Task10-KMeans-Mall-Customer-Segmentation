# Task 10: K-Means Clustering

## Dataset: Mall Customers Dataset

### Objectives

- Perform customer segmentation
- Use Elbow Method
- Determine optimal K value
- Apply K-Means Clustering
- Visualize clusters
- Visualize centroids

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv(
    "Mall_Customers.csv"
)

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
X = df.iloc[:, [3,4]].values
```

# Elbow Method

```python
wcss = []

for i in range(1,11):

    kmeans = KMeans(
        n_clusters=i,
        random_state=42
    )

    kmeans.fit(X)

    wcss.append(
        kmeans.inertia_
    )
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    range(1,11),
    wcss,
    marker="o"
)

plt.title(
    "Elbow Method"
)

plt.xlabel(
    "Number of Clusters"
)

plt.ylabel(
    "WCSS"
)

plt.show()
```

```python
kmeans = KMeans(
    n_clusters=5,
    random_state=42
)

y_kmeans = kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y_kmeans
)

plt.xlabel(
    "Annual Income"
)

plt.ylabel(
    "Spending Score"
)

plt.title(
    "Customer Segments"
)

plt.show()
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X[:,0],
    X[:,1],
    c=y_kmeans
)

plt.scatter(
    kmeans.cluster_centers_[:,0],
    kmeans.cluster_centers_[:,1],
    s=300,
    c="red"
)

plt.xlabel(
    "Annual Income"
)

plt.ylabel(
    "Spending Score"
)

plt.title(
    "Clusters and Centroids"
)

plt.show()
```
